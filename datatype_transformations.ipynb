# Data Transformations


## Transform columns to datetime

Based on the datatypes of the column shown in the df.info() function, it is necessary to transform the format of some of the columns so that they are correctly labelled.

The following columns will be transformed from strings to datetime in the str_to_datetime() function:

- issue_date
- earliest_credit_line
- last_payment_date
- next_payment_date
- last_credit_pull_date

```python
# Transform columns to datetime

df['issue_date'] = pd.to_datetime(df['issue_date'], format ='%b-%y')
df['earliest_credit_line'] = pd.to_datetime(df['earliest_credit_line'], format ='%b-%y')
df['last_payment_date'] = pd.to_datetime(df['last_payment_date'], format ='%b-%y')
df['next_payment_date'] = pd.to_datetime(df['next_payment_date'], format ='%b-%y')
df['last_credit_pull_date'] = pd.to_datetime(df['last_credit_pull_date'], format ='%b-%y')
```

## Tranform columns to category

Based on the low number of unique values shown in df.nunique(), the following columns will be transformed from strings and integers to categories in the str_to_category() function:  

- grade
- term
- sub_grade
- home_ownership
- verification_status
- loan_status
- purpose
- collections_12_mths_ex_med
- employment_length

The columns policy_code and application_type only have one unique value therefore do not contribute meaningful information to the analysis and will be dropped. It will be noted that the application types are all individual and that the policy code for every ID is 1.

```python
# Transform columns to category

columns_to_category = ['verification_status', 'loan_status', 'grade', 'term', 'sub_grade', 'purpose', 'home_ownership', 'collections_12_mths_ex_med', 'employment_length']

df[columns_to_category] = df[columns_to_category].astype('category')
```

## Transform columns to boolean

Based on the prescence of only two unique values in a column, not including NULL values, as shown in df.nunique(), the following columns will be transformed from strings and integers into booleans:

-payment_plan


```python
# Transform columns to boolean

df['payment_plan'] = df['payment_plan'].astype(bool)
```

## Transform columns to numeric

The following columns will be transformed from strings to integers in the str_to_numeric() function:

- loan_amount
- funded_amount
- int_rate
- annual_inc

```python
# Transform columns to numeric

columns_to_numeric= ['loan_amount', 'funded_amount', 'int_rate', 'annual_inc']

def convert_numeric_columns(self, columns_to_numeric):
        for col in columns_to_numeric:
            self.df[col] = pd.to_numeric(self.df[col], errors='coerce')
        return self.df
```
